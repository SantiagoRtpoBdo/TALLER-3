```python
from keras.api._v2.keras import activations
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models

# Carga y preprocesamiento de los datos

# Esta lnea de cdigo utiliza TensorFlow y su mdulo keras para cargar el conjunto de datos MNIST. Este conjunto de datos se divide en dos partes: el conjunto de entrenamiento (train_images) y el conjunto de prueba (que se carga en variables no utilizadas en este fragmento de cdigo).
(train_images, _), (_, _) = tf.keras.datasets.mnist.load_data()
# Esta lnea cambia la forma de las imgenes de entrenamiento. Originalmente, cada imagen es una matriz 2D de 28x28 pxeles, pero es necesario convertirlas en tensores 3D. Se agrega una dimensin adicional al final (1) para indicar que las imgenes son en escala de grises. Se convierten en tipo de datos 'float32' para asegurar que los clculos sean precisos y se pueda realizar el escalado de valores
train_images = train_images.reshape(train_images.shape[0], 28, 28, 1).astype('float32')
# En esta lnea, se realiza el preprocesamiento final. Los valores de pxeles originales estn en el rango de 0 a 255, y esta lnea de cdigo escala esos valores para que estn en el rango de -1 a 1. Esto es importante, ya que las funciones de activacin y las capas de normalizacin suelen funcionar mejor en este rango.
train_images = (train_images - 127.5) / 127.5


# Definicion del generador: El generador es una parte crucial de una GAN y se encarga de generar datos falsos que se asemejen a los datos reales.

def build_generator(latent_dim):
  # Se crea un modelo secuencial vaco para construir la arquitectura del generador en capas secuenciales. Esta conectada con todas las neuronas
  model = models.Sequential()
  # Se agrega una capa densa que toma como entrada el espacio latente y produce una salida con 7x7x128 unidades.
  model.add(layers.Dense(7 * 7 * 128, input_dim=latent_dim))
  # Se aade una funcin de activacin de Leaky ReLU (Rectified Linear Unit con una pendiente pequea para valores negativos) para introducir no linealidad en la red. el alpha es el hipermarametro para esa funcion de activavion
  model.add(layers.LeakyReLU(alpha=0.2))
  # Se utiliza una capa de reorganizacin (Reshape) para transformar la salida de la capa densa en un tensor de forma (7, 7, 128). Esto es necesario para que las capas de convolucin que siguen puedan procesar la informacin adecuadamente.
  model.add(layers.Reshape((7, 7, 128)))
  # Se agrega una capa de convolucin transpuesta que aumenta el tamao del tensor de entrada. Esta capa utiliza un kernel de 4x4, un stride (paso) de (2,2) y utiliza el relleno ('same') para aumentar el tamao del tensor a medida que se procesa.
  model.add(layers.Conv2DTranspose(64, (4, 4), strides=(2,2), padding='same'))
  # Se aplica otra funcin de activacin Leaky ReLU.
  model.add(layers.LeakyReLU(alpha=0.2))
  # Se aade otra capa de convolucin transpuesta que reduce an ms el tamao del tensor y produce la salida final del generador. La funcin de activacin 'tanh' se utiliza para asegurarse de que las salidas estn en el rango de -1 a 1, que es comn para imgenes en escala de grises.
  model.add(layers.Conv2DTranspose(1, (4, 4), strides=(2,2), padding='same', activation = 'tanh'))
  # Finalmente, el modelo se devuelve como resultado de la funcin.
  return model


# Definicion del discriminador: es una red neuronal que se encarga de distinguir entre datos reales y datos generados por el generador.

def build_discriminator():
  # Se crea un modelo secuencial vaco para construir la arquitectura del discriminador en capas secuenciales.
  model = models.Sequential()
  # Se agrega una capa de convolucin 2D que toma imgenes de entrada de 28x28x1 pxeles (escala de grises). Esta capa utiliza un kernel de 3x3, un stride de (2,2) y un relleno ('same'). La capa convolucional se encarga de extraer caractersticas de la imagen.
  model.add(layers.Conv2D(64, (3, 3), strides=(2, 2), padding='same', input_shape=(28, 28, 1)))
  # Se aplica una funcin de activacin de Leaky ReLU para introducir no linealidad en la red.
  model.add(layers.LeakyReLU(alpha=0.2))
  # Se aade una capa de dropout con una tasa de abandono del 40%. El dropout ayuda a evitar el sobreajuste durante el entrenamiento al apagar aleatoriamente algunas neuronas durante cada paso hacia adelante.
  model.add(layers.Dropout(0.4))
  # Se agrega otra capa de convolucin 2D con ms filtros (128 en lugar de 64).
  model.add(layers.Conv2D(128, (3, 3), strides=(2, 2), padding='same'))
  # Se aplica nuevamente una funcin de activacin de Leaky ReLU.
  model.add(layers.LeakyReLU(alpha=0.2))
  # Se aade otra capa de dropout con la misma tasa de abandono del 40%.
  model.add(layers.Dropout(0.4))
  # Se agrega una capa de aplanamiento que convierte el tensor tridimensional en un tensor unidimensional, lo que permite que la informacin se alimente en una capa densa.
  model.add(layers.Flatten())
  # Se aade una capa densa con una sola unidad y una funcin de activacin sigmoide. Esta capa produce una sola salida que representa la probabilidad de que la entrada sea una imagen real (1) o generada por el generador (0).
  model.add(layers.Dense(1, activation='sigmoid'))
  # Finalmente, el modelo se devuelve como resultado de la funcin.
  return model


# Definicion de las dimensiones del espacio latente: El espacio latente es un espacio de nmeros aleatorios a partir del cual el generador de la GAN toma muestras para generar imgenes falsas.
latent_dim = 100


# Construccion y compilacion del descriminador

# Aqu se crea una instancia del discriminador.
discriminator = build_discriminator()
# Se compila.
# loss='binary_crossentropy': Esto establece la funcin de prdida que se utilizar durante el entrenamiento del discriminador. se utiliza la entropa cruzada binaria como funcin de prdida para el discriminador. Esta funcin mide la diferencia entre las salidas reales y las salidas predichas por el discriminador y la compara con la etiqueta real (1 para datos reales y 0 para datos falsos).
# optimizer='adam': Se selecciona el optimizador 'Adam' para ajustar los pesos y minimizar la funcin de prdida.
# metrics=['accuracy']: Se especifica que se debe calcular la precisin (accuracy) como una mtrica durante el entrenamiento del discriminador. Esto proporciona informacin sobre qu tan bien el discriminador est clasificando las imgenes como reales o falsas.
discriminator.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])


# Construccion del generador

# Crea una instancia del generador. La dimensin del espacio latente, se pasa como argumento a la funcin, para construir un generador que pueda muestrear desde ese espacio latente y generar imgenes falsas.
generator = build_generator(latent_dim)


# Construccion de la red GAN: Combina el generador y el discriminador.

# En esta lnea, se establece que el discriminador no debe ser entrenado cuando se entrene la GAN. Esto es importante porque durante el entrenamiento de la GAN, queremos que el discriminador evale las imgenes generadas por el generador, pero no queremos actualizar sus pesos en ese momento.
discriminator.trainable = False
# Se define un tensor de entrada que tiene la misma dimensin que el espacio latente. Esto servir como la entrada para la GAN y ser un tensor de nmeros aleatorios que el generador utilizar para generar imgenes.
gan_input = tf.keras.Input(shape=(latent_dim,))
# Se conecta el tensor de entrada gan_input al generador. Esto significa que, dado un vector de nmeros aleatorios del espacio latente, el generador producir una imagen falsa como salida, que se almacena en la variable x.
x = generator(gan_input)
# Luego, el tensor x (que contiene la imagen falsa generada por el generador) se pasa al discriminador. El discriminador evala esta imagen y produce una salida que representa la probabilidad de que la imagen sea real o falsa.
gan_output =discriminator(x)
# Se crea un modelo gan que conecta el tensor de entrada gan_input con la salida del discriminador gan_output. En esencia, esto crea una GAN que toma vectores aleatorios del espacio latente, genera imgenes falsas y luego el discriminador las evala.
gan = tf.keras.Model(gan_input, gan_output)
# Se compila la GAN utilizando la misma funcin de prdida de entropa cruzada binaria que se utiliza para entrenar el discriminador y el optimizador 'Adam'.
gan.compile(loss='binary_crossentropy', optimizer='adam')


# Entrenamiento

# Este hiperparmetro define el tamao del lote (batch size) utilizado durante el entrenamiento.  Un tamao de lote de 128 significa que se procesarn y actualizarn los pesos de la GAN con 128 imgenes a la vez en cada paso de entrenamiento.
batch_size = 128
# Este hiperparmetro establece el nmero de pocas (iteraciones completas a travs de todo el conjunto de datos) que se utilizarn durante el entrenamiento de la GAN.
epochs = 10000
# Este hiperparmetro controla con qu frecuencia se generan muestras de imgenes falsas durante el entrenamiento. En este caso son los ultimos 3 digitos de mi codigo
sample_interval = 366


# Listas para almacenar las prdidas
d_losses = []
g_losses = []


# Implementa el bucle de entrenamiento para la red GAN.
for epoch in range(epochs):
  # Entrenamiento del discriminador

  # Se selecciona un conjunto aleatorio de ndices de imgenes del conjunto de entrenamiento. Esto se hace para crear un lote de imgenes reales que se utilizarn para entrenar el discriminador.
  idx = np.random.randint(0, train_images.shape[0], batch_size)
  # Se obtienen las imgenes reales correspondientes a los ndices seleccionados en el paso anterior. Estas imgenes se utilizarn para el entrenamiento del discriminador.
  real_images = train_images[idx]
  # Se genera un conjunto de ruido aleatorio del espacio latente.
  noise = np.random.normal(0, 1, (batch_size, latent_dim))
  # Se utiliza el generador para generar imgenes falsas a partir del ruido del espacio latente.
  generated_images = generator.predict(noise)
  # Se crea un conjunto de etiquetas "1" que representan que las imgenes son reales.
  real_labels = np.ones((batch_size, 1))
  # Se crea un conjunto de etiquetas "0" que representan que las imgenes son falsas.
  fake_labels = np.zeros((batch_size, 1))

  # El discriminador se entrena utilizando las imgenes reales y las etiquetas "1" correspondientes.
  d_loss_real = discriminator.train_on_batch(real_images, real_labels)
  #  El discriminador tambin se entrena utilizando las imgenes generadas y las etiquetas "0" correspondientes. Esto se hace para que el discriminador pueda distinguir correctamente entre imgenes reales y falsas.
  d_loss_fake = discriminator.train_on_batch(generated_images, fake_labels)
  # La prdida total del discriminador en esta iteracin se calcula como el promedio entre la prdida en las imgenes reales y la prdida en las imgenes falsas. Esto se hace para seguir la prdida total del discriminador durante el entrenamiento.
  d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)


  # Entrenamiento del generador

  # Se genera un conjunto de ruido aleatorio del espacio latente
  noise = np.random.normal(0, 1, (batch_size, latent_dim))
  #  El generador se entrena utilizando el conjunto de ruido generado (noise) como entrada y las etiquetas "1" (real_labels) como etiquetas de salida.
  g_loss = gan.train_on_batch(noise, real_labels)


  # Registrar las prdidas en cada poca
  d_losses.append(d_loss[0])
  g_losses.append(g_loss)


  if epoch % sample_interval == 0:
    print(f'Epoch {epoch}, D Loss: {d_loss[0]}, G Loss: {g_loss}')
    # Se utiliza el generador para generar imgenes falsas a partir del ruido del espacio latente.
    generated_images = generator.predict(noise)
    # Se realiza una transformacin en las imgenes generadas para ajustar sus valores de pxeles al rango de 0 a 1. Esto es necesario si las imgenes generadas estn en el rango de -1 a 1 para que puedan ser visualizadas correctamente.
    generated_images = 0.5 * generated_images + 0.5
    # mostrar las imgenes generadas.
    fig, axs = plt.subplots(1, 5)

    # Se itera a travs de los primeros 5 elementos de las imgenes generadas
    for i in range(5):
      # Se utiliza una paleta de colores en escala de grises para mostrar las imgenes en blanco y negro.
      axs[i].imshow(generated_images[i, :, :, 0], cmap='gray')
      # Se desactivan los ejes de los subplots, para eliminar los nmeros de los ejes y mejorar la visualizacin.
      axs[i].axis('off')
    plt.show()
```

```python
# Graficar las curvas de prdida
plt.figure(figsize=(12, 6))
plt.plot(range(epochs), d_losses, label='Discriminator Loss')
plt.plot(range(epochs), g_losses, label='Generator Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Curvas de Prdida del Generador y el Discriminador')
plt.show()
```

```python
# Genera una imagen utilizando el generador y un valor en el espacio latente.
def generate_image(generator, latent_dim, random_seed=None):
    if random_seed is not None:
        np.random.seed(random_seed)
    noise = np.random.normal(0, 1, (1, latent_dim))
    generated_image = generator.predict(noise)
    generated_image = 0.5 * generated_image + 0.5  # Ajusta la escala a [0, 1]
    return generated_image[0]

# Genera una imagen aleatoria
random_image = generate_image(generator, latent_dim)

# Muestra la imagen generada
plt.imshow(random_image, cmap='gray')
plt.axis('off')
plt.show()
```
